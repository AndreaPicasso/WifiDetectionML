```python
import csv
import datetime as dt
import dateutil as dtu

csvfile = open('MIE_resources/trainingset.csv')
myReader = csv.reader(csvfile, dialect='excel')
class ProbeRequest:
    
    time = ""
    source = ""
    length = ""
    signal_strength = ""
    sequence_number = ""
    ssid = []
    supp_rates = ""
    ext_supp_rates = ""
    HT_capab_info = ""
    UUID_enrollee = ""
    pdevice_type = ""
    model_name = ""
    model_number = ""
    device_name = ""
    vendor_ext = ""
    vendor_id = ""
    
myProbeRequests = []
for row in myReader:
    
    auxProbe = ProbeRequest()
    auxProbe.time = row[1]
    auxProbe.source = row[2]
    auxProbe.signal_strength = row[3]
    auxProbe.sequence_number = row[4]
    auxProbe.ssid = row[5]
    auxProbe.supp_rates = row[6]
    auxProbe.ext_supp_rates = row[7]
    auxProbe.HT_capab_info = row[8]
    auxProbe.UUID_enrollee = row[14]
    auxProbe.pdevice_type = row[15]
    auxProbe.model_name = row[16]
    auxProbe.model_number = row[17]
    auxProbe.device_name = row[18]
    auxProbe.vendor_ext = row[19]
    auxProbe.vendor_id = row[20]
    myProbeRequests.append(auxProbe)
    
print(len(myProbeRequests))

print(myProbeRequests[5].time)
print(myProbeRequests[5].source)
print(myProbeRequests[5].signal_strength)
print(myProbeRequests[5].sequence_number)
print(myProbeRequests[5].ssid)
print(myProbeRequests[5].supp_rates)
print(myProbeRequests[5].ext_supp_rates)
print(myProbeRequests[5].HT_capab_info)
print(myProbeRequests[5].UUID_enrollee)
print(myProbeRequests[5].pdevice_type)
print(myProbeRequests[5].model_name)
print(myProbeRequests[5].model_number)
print(myProbeRequests[5].device_name)
print(myProbeRequests[5].vendor_ext)
print(myProbeRequests[5].vendor_id)

for i in range(1, len(myProbeRequests)):
    myProbeRequests[i-1] = myProbeRequests[i]
    
myProbeRequests.pop()
    
for i in range(0, len(myProbeRequests)):
    myProbeRequests[i].time = dtu.parser.parse(myProbeRequests[i].time)

print(myProbeRequests[len(myProbeRequests)-1].time)
print(myProbeRequests[0].time)
diff_time = myProbeRequests[len(myProbeRequests)-1].time - myProbeRequests[0].time
print(type(diff_time))
print(diff_time)

```

```python
def distance(p1, p2):
    
    distance = 0
    
    w_time = 0.1
    w_source = 1
    w_sig_str = 1
    w_seq_num = 1
    w_ssid = 10
    w_supp_rates = 10
    w_ext_supp_rates = 100
    w_HT_capab_info = 100
    w_UUID_enrollee = 100
    w_pdevice_type = 100
    w_model_name = 100
    w_model_number = 100
    w_device_name = 100
    w_vendor_ext = 1
    w_vendor_id = 1
    
    diff_time = abs(p1.time - p2.time)
    diff_time = diff_time.total_seconds()
    distance += diff_time * w_time
    
    if p1.source != p2.source:
        distance += w_source
        
    distance += abs(p1.signal_strength - p2.signal_strength) * w_sig_str
    distance += abs(p1.sequence_number - p2.sequence_number) * w_seq_num
    
    if p1.ssid != p2.ssid:
        distance += w_ssid
    # aumenta se sono diverse? io farei avvicinare se sono uguali
        
    if p1.supp_rates != p2.supp_rates:
        distance += w_supp_rates
        
    if p1.ext_supp_rates != p2.ext_supp_rates:
        distance += w_ext_supp_rates
        
    if p1.HT_capab_info != p2.HT_capab_info:
        distance += w_HT_capab_info
        
    if p1.UUID_enrollee != p2.UUID_enrollee:
        distance += w_UUID_enrollee
        
    if p1.pdevice_type != p2.pdevice_type:
        distance += w_pdevice_type
        
    if p1.model_name != p2.model_name:
        distance += w_model_name
        
    if p1.model_number != p2.model_number:
        distance += w_model_number
            
    if p1.device_name != p2.device_name:
        distance += w_device_name
            
    if p1.vendor_ext != p2.vendor_ext:
        distance += w_vendor_ext
            
    if p1.vendor_id != p2.vendor_id:
        distance += w_vendor_id
        
    return distance
```

```python
def vectorize(myList):
    
    w_time = 0.1
    w_source = 1
    w_sig_str = 1
    w_seq_num = 1
    w_ssid = 10
    w_supp_rates = 10
    w_ext_supp_rates = 100
    w_HT_capab_info = 100
    w_UUID_enrollee = 100
    w_pdevice_type = 100
    w_model_name = 100
    w_model_number = 100
    w_device_name = 100
    w_vendor_ext = 1
    w_vendor_id = 1
    
    vectorizedList = []
    origin = []
    origin.append(0)
    origin.append(0)
    origin.append(w_sig_str * abs(int(myList[0].signal_strength)))
    origin.append(w_seq_num * int(myList[0].sequence_number))
    origin.append(0)
    origin.append(0)
    origin.append(0)
    origin.append(0)
    origin.append(0)
    origin.append(0)
    origin.append(0)
    origin.append(0)
    origin.append(0)
    origin.append(0)
    origin.append(0)
    vectorizedList.append(origin);
    
    for i in range(1, len(myList)):
        
        source_dist = 0
        ssid_dist = 0
        supp_rates_dist = 0
        ext_supp_rates_dist = 0
        HT_capab_info_dist = 0
        UUID_enrollee_dist = 0
        pdevice_type_dist = 0
        model_name_dist = 0
        model_number_dist = 0
        device_name_dist = 0
        vendor_ext_dist = 0
        vendor_id_dist = 0
        
        aux_vec = []
        aux_vec.append(w_time * (myList[i].time - myList[0].time).total_seconds())
        
        if myList[i-1].source != myList[i].source:
            source_dist = vectorizedList[i-1][1] + w_source
            
        for k in range(i-1, 0):
            
            if myList[i].source == myList[k].source:
                source_dist = vectorizedList[k][1]
            
        aux_vec.append(source_dist);
        
        aux_vec.append(w_sig_str * abs(int(myList[i].signal_strength)))
        aux_vec.append(w_seq_num * int(myList[i].sequence_number))
        
        if myList[i-1].ssid != myList[i].ssid:
            ssid_dist = vectorizedList[i-1][4] + w_ssid
            
        if myList[i-1].supp_rates != myList[i].supp_rates:
            supp_rates_dist = vectorizedList[i-1][5] + w_supp_rates
            
        if myList[i-1].ext_supp_rates != myList[i].ext_supp_rates:
            ext_supp_rates_dist = vectorizedList[i-1][6] + w_ext_supp_rates
            
        if myList[i-1].HT_capab_info != myList[i].HT_capab_info:
            HT_capab_info_dist = vectorizedList[i-1][7] + w_HT_capab_info
            
        if myList[i-1].UUID_enrollee != myList[i].UUID_enrollee:
            UUID_enrollee_dist = vectorizedList[i-1][8] + w_UUID_enrollee
            
        if myList[i-1].pdevice_type != myList[i].pdevice_type:
            pdevice_type_dist = vectorizedList[i-1][9] + w_pdevice_type
            
        if myList[i-1].model_name != myList[i].model_name:
            model_name_dist = vectorizedList[i-1][10] + w_model_name
            
        if myList[i-1].model_number != myList[i].model_number:
            model_number_dist = vectorizedList[i-1][11] + w_model_number
            
        if myList[i-1].device_name != myList[i].device_name:
            device_name_dist = vectorizedList[i-1][12] + w_device_name
            
        if myList[i-1].vendor_ext != myList[i].vendor_ext:
            vendor_ext_dist = vectorizedList[i-1][13] + w_vendor_ext
            
        if myList[i-1].vendor_id != myList[i].vendor_id:
            vendor_id_dist = vectorizedList[i-1][14] + w_vendor_id
            
        for k in range(i-1, 0):
                
            if myList[i].ssid == myList[k].ssid:
                ssid_dist = vectorizedList[k][4]
                
            if myList[i].supp_rates == myList[k].supp_rates:
                supp_rates_dist = vectorizedList[k][5]
                
            if myList[i].ext_supp_rates == myList[k].ext_supp_rates:
                ext_supp_rates_dist = vectorizedList[k][6]
                
            if myList[i].HT_capab_info == myList[k].HT_capab_info:
                HT_capab_info_dist = vectorizedList[k][7]
                
            if myList[i].UUID_enrollee == myList[k].UUID_enrollee:
                UUID_enrollee_dist = vectorizedList[k][8]
                
            if myList[i].pdevice_type == myList[k].pdevice_type:
                pdevice_type_dist = vectorizedList[k][9]
                
            if myList[i].model_name == myList[k].model_name:
                model_name_dist = vectorizedList[k][10]
                
            if myList[i].model_number == myList[k].model_number:
                model_number_dist = vectorizedList[k][11]
            
            if myList[i].device_name == myList[k].device_name:
                device_name_dist = vectorizedList[k][12]
            
            if myList[i].vendor_ext == myList[k].vendor_ext:
                vendor_ext_dist = vectorizedList[k][13]
            
            if myList[i].vendor_id == myList[k].vendor_id:
                vendor_id_dist = vectorizedList[k][14]
            
            
        aux_vec.append(ssid_dist);
        aux_vec.append(supp_rates_dist);
        aux_vec.append(ext_supp_rates_dist);
        aux_vec.append(HT_capab_info_dist);
        aux_vec.append(UUID_enrollee_dist);
        aux_vec.append(pdevice_type_dist);
        aux_vec.append(model_name_dist);
        aux_vec.append(model_number_dist);
        aux_vec.append(device_name_dist);
        aux_vec.append(vendor_ext_dist);
        aux_vec.append(vendor_id_dist);
        
        vectorizedList.append(aux_vec)
        
    return vectorizedList
        
```
